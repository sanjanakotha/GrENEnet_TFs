```python
import pandas as pd
```

```python
pred_df = pd.read_csv("06_all_prediction_intervals.csv", index_col = 0)
pred_df
```

```python
measured_df = pd.read_csv("10_padi_annotated_ADs.csv", index_col = 0)
measured_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_context('talk')

# Merge predicted and measured ADs into a single dataframe with a label
pred_df['type'] = 'predicted'
measured_df['type'] = 'measured'
measured_df['predictor'] = 'Annotated'

combined = pd.concat([
    pred_df.rename(columns={'Start':'start','End':'end'}),
    measured_df
], ignore_index=True)

# Get a consistent y-axis order across all plots
predictor_order = list(combined['predictor'].unique())
predictor_to_y = {pred: i for i, pred in enumerate(predictor_order)}

# Example plotting per tair_ID
for tair in measured_df['tair_ID'].unique():
    df_plot = combined[combined['tair_ID'] == tair]
    plt.figure(figsize=(8, 2))
    
    for _, row in df_plot.iterrows():
        y = predictor_to_y[row['predictor']]  # map predictor name to numeric
        plt.hlines(y=y, xmin=row['start'], xmax=row['end'],
                   colors='red' if row['type']=='measured' else 'blue',
                   lw=6, alpha=0.7)
    
    plt.title(tair)
    plt.xlabel('Amino acid position')
    plt.ylabel('Predictor')
    
    # Set ticks and labels
    plt.yticks(ticks=list(predictor_to_y.values()), labels=list(predictor_to_y.keys()))
    plt.ylim(-1, len(predictor_order))
    
    sns.despine()
    plt.show()
```
