```python
import pandas as pd
import protfasta
```

```python
seqs = protfasta.read_fasta("01_tair_arabidopsis_raw.fasta",
                            expect_unique_header = False,
                           duplicate_record_action = 'ignore',
                           invalid_sequence_action = 'ignore')
seqs = pd.DataFrame({"id" : seqs.keys(), "seq" : seqs.values()})
seqs
```

```python
seqs["gene"] = seqs["id"].str.split("Symbols: ").str[1].str.split("|").str[0].str[:-1]
seqs
```

```python
seqs["tair_ID"] = seqs["id"].str.split("|").str[0].str[:-1]
seqs
```

```python
seqs["seq_cleaned"] = seqs["seq"].str[:-1]
seqs
```

```python
seqs[seqs["seq_cleaned"].str.contains("\*")]
```

```python
protfasta.write_fasta(dict(zip(seqs["gene"], seqs["seq_cleaned"])), "01_tair_arabidopsis_cleaned.fasta")
```

```python
seqs.to_csv("01d_tair_arabidopsis_cleaned.csv")
```
